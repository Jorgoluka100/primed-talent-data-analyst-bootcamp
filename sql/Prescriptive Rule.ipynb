```python
import math

def calculate_drivers(total_packages, hours, rate_per_hour, buffer, min_d, max_d):
    # Step 1 & 2
    base_drivers = (total_packages / hours) / rate_per_hour
    # Step 3
    with_buffer = base_drivers * (1 + buffer)
    # Step 4
    final_count = math.ceil(with_buffer)
    # Step 5
    clamped_count = max(min_d, min(final_count, max_d))
    
    return clamped_count

# Usage
recommended = calculate_drivers(450, 4, 8, 0.15, 5, 25)
print(f"Recommended Drivers: {recommended}")
```

```python
#Exercise 2 (Intermediate): Building a Prescriptive Rule

def check_churn_risk(customer_id, calls_this_month):
    # Predictive Model Constants
    intercept = 0.25
    coefficient = -0.05
    target_churn = 0.15
    
    # Prediction
    predicted_churn = intercept + (coefficient * calls_this_month)
    
    # Prescriptive Logic
    if predicted_churn >= target_churn:
        return {
            "status": "AT RISK",
            "predicted_churn": f"{predicted_churn:.1%}",
            "action": "Trigger Proactive Retention Offer"
        }
    else:
        return {
            "status": "HEALTHY",
            "predicted_churn": f"{predicted_churn:.1%}",
            "action": "No Action Required"
        }

# Example usage:
print(check_churn_risk("CUST_123", calls_this_month=1))
```
